```python
from mpl_toolkits.mplot3d import Axes3D 
import matplotlib.pyplot as plt
import numpy as np
```

Cabeceras necesarias para gráficas en 3D

```python
#get_ipython().magic(u'matplotlib inline')
%matplotlib inline
```

Para poder graficar aquí mismo

```python
x = np.arange(-10,10,0.25)
y = np.arange(-10,10,0.25)
x, y = np.meshgrid(x, y)
```

Arreglos de puntos (x,y) unido en una rejilla

```python
z = x**2-y**2
```

Función a graficar

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

```python
ax.plot_surface(x, y, z,  rstride=4, cstride=4, color='b')
cset = ax.contour(x, y, z, zdir='z', offset=-100)
cset = ax.contour(x, y, z, zdir='x', offset=-10)
cset = ax.contour(x, y, z, zdir='y', offset=10)
```

Se hace la gráfica y las curvas de nivel

```python
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```
